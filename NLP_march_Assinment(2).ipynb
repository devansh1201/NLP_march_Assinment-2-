```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
#uploding text data
```

```python
text_file="Chess is a strategic board game played between two players.Each player begins with sixteen pieces: one king, one queen, two rooks, two bishops, two knights, and eight pawns.he main objective is simple: checkmate the opponent’s king!Players must think carefully before every move, because one mistake can change the entire game.Chess improves concentration, patience, and problem-solving skills."
```

```python
print(text_file)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
# punctuation removing
tokens = word_tokenize(text_file)
tokens_punctuation=[word for word in tokens if word not in string.punctuation]
print("\nafter removing punctuation:-")
print(tokens_punctuation)
```

```python
import nltk
nltk.download('stopwords')
```

```python
#stop words removal
stop_words = set(stopwords.words('english'))
filtered_words=[w for w in tokens_punctuation if w.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_words=[stemmer.stem(word) for word in filtered_words]

print(stemmed_words)
```

```python
#Lemmatization

import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in filtered_words]

print(lemmatized_words)
```

```python
#feture extrction process
```

```python
string_converted_data=" ".join(lemmatized_words)
print(string_converted_data)
```

```python
#first tecniqe bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus=[string_converted_data]#corpus object assined string text data
vectorizer=CountVectorizer()
bag_of_words=vectorizer.fit_transform(corpus)
# each word freqvency how many times it appiyer in the text
print(bag_of_words.toarray())
print(vectorizer.get_feature_names_out())
```

```python
#second tecnique TF-IDF
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus=[string_converted_data]
vectorizer1=TfidfVectorizer()
tfidf_vac=vectorizer1.fit_transform(corpus)
print(tfidf_vac.toarray())
print(vectorizer1.get_feature_names_out())
```

```python
text_sentences = [
    ["chess", "is", "a", "strategic", "board", "game", "played", "between", "two", "players"],
    ["each", "player", "begins", "with", "sixteen", "pieces", "one", "king", "one", "queen", "two", "rooks", "two", "bishops", "two", "knights", "and", "eight", "pawns"],
    ["the", "main", "objective", "is", "simple", "checkmate", "the", "opponent’s", "king"],
    ["players", "must", "think", "carefully", "before", "every", "move", "because", "one", "mistake", "can", "change", "the", "entire", "game"],
    ["chess", "improves", "concentration", "patience", "and", "problem-solving", "skills"]
]
```

```python
from gensim.models import Word2Vec
model=Word2Vec(text_sentences,min_count=1,window=5,workers=4)
```

```python
print('print vector for game')
print(model.wv['game'])#inbield created object wv for Word2Vec
print('\nWord similar to game ')
print(model.wv.most_similar('game'))#show similarity score
```

```python
# traying to increasing similarity score
from gensim.models import Word2Vec
model=Word2Vec(text_sentences,min_count=1,window=5,epochs=200,vector_size=100)
```

```python
# checking player word similarity word
print('print vector for player')
print(model.wv['player'])
print('\nWord similar to player')
print(model.wv.most_similar('player'))
#now our similarity score incresed.
```

```python
#checking cosine similarity
```

```python
import numpy as np
def sent_vec(sentence):
  word=sentence.lower().split()#creating text data into lover case
  vectors=[]
  for word in word:
    if word in model.wv:
      vectors.append(model.wv[word])
  return np.mean(vectors,axis=0)

sentence="Chess is a strategic board game played between two players"
vec=sent_vec(sentence)
print("sentence vector:")
print(vec)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sentence_1=sent_vec("Chess is a strategic board game played between two players.")
sentence_2=sent_vec("Chess improves concentration, patience, and problem-solving skills.")
similarity= cosine_similarity([sentence_1],[sentence_2])
print("sentence_similarity:\n")
print(similarity[0][0])#cosine similarity score
```

```python
from sklearn.linear_model import LogisticRegression
#creating X and y with sentence data.
X=[
sent_vec("The main objective is simple: checkmate the opponent’s king!"),
sent_vec("Chess improves concentration, patience, and problem-solving skills."),
sent_vec("Chess is a strategic board game played between two players."),
sent_vec("Chess learning is boaring"),
    sent_vec("chess learning is very hard and not a good mind game"),
    sent_vec("always playing chess is time consuming")


]
y=[1,1,1,0,0,0]
model_1= LogisticRegression()
#Applying logistic regression
model_1.fit(X,y)
test=sent_vec("i want to play chess")
#compeyaring to other sentences negetive sentence is more like this sentence so we found [0] in out-put.
print("pridiction", model_1.predict([test]))
```

```python
from sklearn.cluster import KMeans
X=[sent_vec("The main objective is simple: checkmate the opponent’s king!"),
sent_vec("Chess improves concentration, patience, and problem-solving skills."),
sent_vec("Chess is a strategic board game played between two players."),
sent_vec("Chess learning is boaring")
]
kmeans=KMeans(n_clusters=2)
kmeans.fit(X)
print("clusters",kmeans.labels_)
#showing first sentence as first cluster and other take as second cluster
```

```python
#checking for three clusters
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
print("clusters",kmeans.labels_)
#taking first and third as one cluster,second as second cluster ,forth as third cluster
```
